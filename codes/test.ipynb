```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

xs = list(range(1, 100))
ys = list(range(1, 100))
fig, ax = plt.subplots()
plt.ylabel("Loss values")
plt.xlabel('Num. iteration')
x, y = [], []

def update(i):
    print(i)
    x.append(xs[i])
    y.append(ys[i])
    ax.plot(x, y, color='blue', label='Loss value')
animator = animation.FuncAnimation(fig, update,
                                    interval=50, repeat=False)
animator.save('a.gif')
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

xs1 = list(range(1, 100))
ys1 = np.loadtxt("loss1.txt", dtype=float)
xs2 = list(range(1, 100))
ys2 = np.loadtxt("loss2.txt", dtype=float)
xs3 = list(range(1, 100))
ys3 = np.loadtxt("loss3.txt", dtype=float)
xs4 = list(range(1, 100))
ys4 = np.loadtxt("loss4.txt", dtype=float)

fig, axs = plt.subplots(2, 2)
plt.ylabel("Loss values")
plt.xlabel('Num. iteration')
plt.xlim(0, 100)
plt.ylim(-0.05, 0.5)
x, y1, y2 = [], [], []

def update(i):
    x.append(xs1[i])
    y1.append(ys1[i])
    y2.append(ys2[i])
    if i == 0:  # Add label only in the fourth frame
        axs[0, 0].plot(x, y1, color='blue', label='Loss value 1')
    else:
        axs[0, 0].plot(x, y1, color='blue')
    if i == 0:  # Add label only in the fourth frame
        axs[0, 1].plot(x, y2, color='red', label='Loss value 2')
    else:
        axs[0, 1].plot(x, y2, color='red')



animator1 = animation.FuncAnimation(fig, update, frames=len(xs1), interval=50, repeat=False)
animator1.save('a.gif')
```
