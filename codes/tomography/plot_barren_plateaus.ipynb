```python
from matplotlib import pyplot as plt
num_qubits = 5
num_layers = range(1, 6)
# graph-star ansatz
wchain_variances = [
    2.266503073665902080e-05,
    2.397433141862732420e-04,
    4.911772668003688158e-05,
    4.084392499335541153e-05,
    1.006935116111977117e-05
]

walternating_variances = [
    3.352270956349658287e-04,
    7.099164453061283407e-04,
    1.164318023671071872e-05,
    2.159793146693802471e-04,
    1.069792861943279005e-04
]

walltoall_variances = [
    3.135691344528749734e-05,
    9.375298504597546508e-06,
    2.046162691969384431e-05,
    1.353840516685076005e-05,
    6.282004015683642843e-06

]


wchain_variances = np.array(wchain_variances)
walternating_variances = np.array(walternating_variances)
walltoall_variances = np.array(walltoall_variances)
num_layers = np.array(num_layers)
p_star = np.polyfit(num_layers, np.log(wchain_variances), 1)
p_polygon = np.polyfit(num_layers, np.log(walternating_variances), 1)
p_linear = np.polyfit(num_layers, np.log(walltoall_variances), 1)
print(p_star[0])
print(p_polygon[0])
print(p_linear[0])
plt.plot(num_layers, wchain_variances, marker='o',
         linestyle='dotted', color='red', label = 'Wchain')
plt.plot(num_layers, walternating_variances, marker='o',
         linestyle='dotted', color='green', label = 'Walternating')
plt.plot(num_layers, walltoall_variances, marker='o',
         linestyle='dotted', color='blue', label = 'Walltoall')
# plt.semilogy(num_layers, np.exp(p_star[0] * num_layers + p_star[1]), linestyle='dashed',
#              color='red', label="Graph-star ansatz. Slope {:3.2f}".format(p_star[0]))
# plt.semilogy(num_layers, np.exp(p_polygon[0] * num_layers + p_polygon[1]), linestyle='dashed',
#              color='green', label="Graph-polygon ansatz. Slope {:3.2f}".format(p_polygon[0]))
# plt.semilogy(num_layers, np.exp(p_linear[0] * num_layers + p_linear[1]), linestyle='dashed',
#              color='blue', label="Linear ansatz. Slope {:3.2f}".format(p_linear[0]))
plt.xticks(num_layers)
plt.legend()
# plt.savefig('bp.eps', format='eps', dpi=1000)
```
