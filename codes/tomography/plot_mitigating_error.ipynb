```python
import qiskit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
noises = [0.01, 0.02, 0.03, 0.04]
color = ["orange", "blue", "cyan", "red"]
i = 0
for noise in noises:
    loss = pd.read_csv("./loss_values" + str(noise) + ".csv", sep=",", header=None)
    plt.plot(loss, color = color[i], label = noise)
    loss_mitigating = pd.read_csv("./loss_values" + str(noise) + "mitigating.csv", sep=",", header=None)
    plt.plot(loss_mitigating, color = color[i], linestyle = "dotted", label = noise)
    i = i + 1
plt.legend()
```
