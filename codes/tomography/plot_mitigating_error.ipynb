```python
noises = [0.01, 0.02, 0.03, 0.04]
for noise in noises:
    loss = pd.read_csv("./loss_values" + str(noise) + ".csv", sep=",", header=None)
    plt.plot(loss, label = noise)
plt.legend()
```
