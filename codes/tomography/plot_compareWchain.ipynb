```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
fidelitiess = []

for i in range(1, 6):
    path = "../../experiments/tomography_wchain_" + str(i) + '/' + str(3) + '/fidelities.csv'

    fidelities = pd.read_csv(path, sep=",", header=None)
    fidelities = fidelities.applymap(lambda s: complex(s.replace('i', 'j'))).values
    fidelitiess.append(np.max(fidelities))
plt.plot(range(1, 6), fidelitiess, label = str(i) + ' layer', marker = 'o')

plt.xlabel('Number of layers')
plt.xticks(np.arange(1, 6, 1))
plt.yticks(np.arange(0, 1.4, 0.2))
plt.ylabel('Value')
plt.title('Fidelity')
plt.legend()
plt.show()
```
