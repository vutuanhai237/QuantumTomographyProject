```python
tracess = []
qubits = [3,4,5]
for j in qubits:
    for i in range(1, 6):
        path = "../../experiments/tomography_walternating_" + str(i) + '/' + str(j) + '/traces_qng.csv'

        traces = pd.read_csv(path, sep=",", header=None)
        # fidelities = fidelities.applymap(lambda s: complex(s.replace('i', 'j'))).values
        tracess.append(np.min(traces))
    plt.plot(range(1, 6), tracess, label = str(j) + ' qubits', marker = 'o')
    tracess = []

plt.xlabel('Number of layers')
plt.xticks(np.arange(1, 6, 1))
plt.yticks(np.arange(0, 1.4, 0.2))
plt.ylabel('Value')
plt.title('Traces')
plt.legend()
plt.show()
```

```python
fidelitiess = []
qubits = [3,4,5]
for j in qubits:
    for i in range(1, 6):
        path = "../../experiments/tomography_walternating_" + str(i) + '/' + str(j) + '/fidelities_qng.csv'

        fidelities = pd.read_csv(path, sep=",", header=None)
        fidelities = fidelities.applymap(lambda s: complex(s.replace('i', 'j'))).values
        fidelitiess.append(np.max(fidelities))
    plt.plot(range(1, 6), fidelitiess, label = str(j) + ' qubits', marker = 'o')
    fidelitiess = []

plt.xlabel('Number of layers')
plt.xticks(np.arange(1, 6, 1))
plt.yticks(np.arange(0, 1.4, 0.2))
plt.ylabel('Value')
plt.title('Fidelity')
plt.legend()
plt.show()
```
