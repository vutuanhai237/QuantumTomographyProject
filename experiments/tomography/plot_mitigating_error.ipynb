```python
import qiskit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
noises = [0.01, 0.02, 0.03, 0.04]
color = ["orange", "blue", "cyan", "red"]
i = 0
for noise in noises:
    loss = pd.read_csv("./noise_qng/loss_values" + str(noise) + ".csv", sep=",", header=None)
    plt.plot(loss, color = color[i], label = noise)
    loss_mitigating = pd.read_csv("./noise_qng/loss_values" + str(noise) + "mitigating.csv", sep=",", header=None)
    plt.plot(loss_mitigating, color = color[i], linestyle = "dotted", label = noise)
    i = i + 1
plt.legend()
```

```python
free_noise_log = np.sqrt(pd.read_csv("../../experiments/tomography/tomography_wchain_2/5/loss_values_adam.csv", sep=",", header=None))

noises = [0.0, 0.01, 0.02, 0.03, 0.04]
losss = [np.min(free_noise_log)]
losss_mitigating = [np.min(free_noise_log)]
i = 0
for noise in noises:
    if noise != 0.0:
        loss = pd.read_csv("./noise_adam/loss_values" + str(noise) + ".csv", sep=",", header=None)
        losss.append(np.min(loss))
        loss_mitigating = pd.read_csv("./noise_adam/loss_values" + str(noise) + "mitigating.csv", sep=",", header=None)
        losss_mitigating.append(np.min(loss_mitigating))
plt.plot(noises, losss, marker = 'o', color = 'purple', linestyle = 'dotted', label = 'noise without mitigation')
plt.plot(noises, losss_mitigating, color = 'cyan', marker = 'o', label = 'noise with mitigation')
plt.xticks([0.0, 0.01, 0.02, 0.03, 0.04])
plt.ylabel('Fubini-study distance')
plt.xlabel('$\epsilon$')
plt.legend()
```
